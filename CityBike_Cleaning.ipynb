```python
df_2018 = pd.DataFrame()

for month in ["01", "02","03","04","05","06","07","08","09","10","11","12"]:
    filepath = f"../../OneDrive/Desktop/CitiBike_Data/2018/JC-2018{month}-citibike-tripdata.csv"
    
    df_input = pd.read_csv(filepath, low_memory=False)
    sample = df_input.sample(frac=0.5, random_state=42)
    df_2018 = df_2018.append(sample)
    
    print(month)
```

```python
df_2018.reset_index(drop=True, inplace=True)
df_2018
```

```python
df_2018.to_csv("2018sample.csv", index=False)
```
